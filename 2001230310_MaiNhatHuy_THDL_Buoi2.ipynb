```
import pandas as pd

df = pd.read_csv("dulieuxettuyendaihoc.csv")
df.head()
```

```
dh1_sorted = df.sort_values(by="DH1")
dh1_sorted.head()
```

```
dh2_sorted_by_gender = df.sort_values(by=["GT", "DH2"])
dh2_sorted_by_gender.head()
```

```
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.50)

def q3(x):
    return x.quantile(0.75)
pivot_KT = pd.pivot_table(
    df,
    values="DH1",
    index="KT",
    aggfunc=["count", "sum", "mean", "median", "min", "max", "std", q1, q2, q3]
)

pivot_KT
```

```
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.50)

def q3(x):
    return x.quantile(0.75)
pivot_KT_KV = pd.pivot_table(
    df,
    values="DH1",
    index=["KT", "KV"],
    aggfunc=["count", "sum", "mean", "median", "min", "max", "std", q1, q2, q3]
)

pivot_KT_KV
```

```
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.50)

def q3(x):
    return x.quantile(0.75)
pivot_KT_KV_DT = pd.pivot_table(
    df,
    values="DH1",
    index=["KT", "KV", "DT"],
    aggfunc=["count", "sum", "mean", "median", "min", "max", q1, q2, q3]
)

pivot_KT_KV_DT
```

```
#PHẦN 2: TRÌNH BÀY DỮ LIỆU
gt_freq = df["GT"].value_counts().to_frame(name="Tan_so")
gt_freq["Tan_suat"] = gt_freq["Tan_so"] / gt_freq["Tan_so"].sum()
gt_freq
```

```
import matplotlib.pyplot as plt
gt_freq["Tan_so"].plot(kind="bar")
plt.title("Bieu do tan so Gioi tinh")
plt.xlabel("Gioi tinh")
plt.ylabel("Tan so")
plt.show()
```

```
gt_freq["Tan_suat"].plot(kind="pie", autopct="%.1f%%")
plt.title("Bieu do tan suat Gioi tinh")
plt.ylabel("")
plt.show()
```

```
gt_freq_sorted = gt_freq.sort_index()
gt_freq_sorted["Tan_suat_tich_luy"] = gt_freq_sorted["Tan_suat"].cumsum()

plt.plot(gt_freq_sorted.index, gt_freq_sorted["Tan_suat_tich_luy"], marker="o")
plt.title("Da giac tich luy tan suat Gioi tinh")
plt.xlabel("Gioi tinh")
plt.ylabel("Tan suat tich luy")
plt.ylim(0,1)
plt.grid()
plt.show()
```

```
df["US_TBM1"] = df[["T1","L1","H1","S1","V1","X1","D1","N1"]].mean(axis=1)
df["US_TBM2"] = df[["T2","L2","H2","S2","V2","X2","D2","N2"]].mean(axis=1)
df["US_TBM3"] = df[["T3","L3","H3","S3","V3","X3","D3","N3"]].mean(axis=1)
for col in ["US_TBM1", "US_TBM2", "US_TBM3"]:
    print(f"\nThong ke mo ta {col}")
    display(df[col].describe())

    df[col].plot(kind="hist", bins=10)
    plt.title(f"Bieu do phan bo {col}")
    plt.xlabel(col)
    plt.show()
```

```
df_nam = df[df["GT"] == "M"]

dt_nam_freq = df_nam["DT"].value_counts().to_frame(name="Tan_so")
dt_nam_freq["Tan_suat"] = dt_nam_freq["Tan_so"] / dt_nam_freq["Tan_so"].sum()
dt_nam_freq
dt_nam_freq["Tan_so"].plot(kind="bar")
plt.title("Tan so DT cua hoc sinh nam")
plt.xlabel("DT")
plt.ylabel("Tan so")
plt.show()
```

```
df_dk = df[
    (df["GT"] == "M") &
    (df["DT"] == 1) &
    (df["DH1"] >= 5.0) &
    (df["DH2"] >= 4.0) &
    (df["DH3"] >= 4.0)
]
if df_dk.empty:
    print("Khong co hoc sinh nao thoa man dieu kien.")
else:
    kv_freq = df_dk["KV"].value_counts().to_frame(name="Tan_so")
    kv_freq["Tan_suat"] = kv_freq["Tan_so"] / kv_freq["Tan_so"].sum()

    kv_freq["Tan_so"].plot(kind="bar")
    plt.title("Phan bo KV (Nam, Kinh, du dieu kien)")
    plt.xlabel("KV")
    plt.ylabel("Tan so")
    plt.show()
```

```
df_2nt = df[
    (df["KV"] == "2NT") &
    (df["DH1"] >= 5.0) &
    (df["DH2"] >= 5.0) &
    (df["DH3"] >= 5.0)
]

df_2nt[["DH1","DH2","DH3"]].describe()
df_2nt[["DH1","DH2","DH3"]].plot(kind="box")
plt.title("Bieu do boxplot DH1, DH2, DH3 (KV=2NT, >=5)")
plt.show()
```

```
#PHẦN 3: Trực quan hóa dữ liệu theo nhóm phân loại
bang = df.groupby(['KT', 'KV']).size().unstack()

# In bảng
print(bang)

# Vẽ biểu đồ
bang.plot(kind='bar')

plt.title("Số lượng thí sinh theo KV và KT")

plt.xlabel("Khối thi")
plt.ylabel("Số lượng")

plt.legend(title='Khu vực')

plt.show()
```

```
#PHẦN 4: Trực quan hóa dữ liệu nâng cao

plt.plot(df['T1'])

plt.title("Simple Line T1")
plt.show()


def xep_loai(diem):
    if diem < 5:
        return 'k'
    elif diem < 7:
        return 'tb'
    elif diem < 8:
        return 'kha'
    else:
        return 'g'

df['phanloaiT1'] = df['T1'].apply(xep_loai)


tan_so = df['phanloaiT1'].value_counts()

print(tan_so)


bang = df.groupby(['T1', 'phanloaiT1']).size().unstack(fill_value=0)

bang.plot()

plt.title("Multiple Line")
plt.show()


plt.stem(df.index,
         df['T1'])

plt.title("Drop-line T1")

plt.show()
```
